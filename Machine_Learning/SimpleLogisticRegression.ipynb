# Logistic Regression Implementation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
from sklearn.datasets import load_iris
dataset = load_iris()
```

```python
df = pd.DataFrame(dataset.data, columns = dataset.feature_names)
df['target'] = dataset.target
df.head()
```

```python
df_copy = df[df['target']!=2]
```

```python
df_copy.info()
```

```python
df.describe()
```

```python
# Spliting the dataset into dependent and independent features
x = df.iloc[:,:-1]
y = df['target']
```

```python
# Further spliting the dataset into train and test data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size =0.20,random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
logistic = LogisticRegression()
```

```python
logistic.fit(x_train,y_train)
```

```python
y_pred = logistic.predict(x_test)
```

```python
# Accuracy Checking
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
print(confusion_matrix(y_pred,y_test))
```

```python
print(accuracy_score(y_pred,y_test))
```

```python
print(classification_report(y_pred,y_test))
```

```python
# Hyperparameter tuning
from sklearn.model_selection import GridSearchCV
```

```python
parameter = {'penalty':['l1','l2','elasticnet',None],'C':[1,10,20]}
```

```python
clf = GridSearchCV(logistic,param_grid= parameter,cv = 5)
```

```python
clf.fit(x_train,y_train)
```

```python
clf.best_params_
```

```python
y_pred = clf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
print(confusion_matrix(y_pred,y_test))
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
rcf = RandomizedSearchCV(logistic,param_distributions= parameter)
```

```python
rcf.fit(x_train,y_train)
```

```python
rcf.best_params_
```

```python
y_pred = rcf.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
print(confusion_matrix(y_pred,y_test))
print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
```
