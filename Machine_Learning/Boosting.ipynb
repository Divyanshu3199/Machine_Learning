# Classification

```python
from sklearn.datasets import make_classification
```

```python
x,y = make_classification(n_samples = 1000,n_features =4,n_informative = 2, n_redundant= 0,random_state = 42,shuffle = False)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.33, random_state = 42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.ensemble import AdaBoostClassifier
```

```python
abc = AdaBoostClassifier()
```

```python
abc.fit(x_train,y_train)
```

```python
y_pred1 = abc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred1,y_test)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gbc = GradientBoostingClassifier()
```

```python
gbc.fit(x_train,y_train)
```

```python
y_pred2 = gbc.predict(x_test)
```

```python
print(accuracy_score(y_pred2,y_test))
```

```python
xgbc.fit(x_train,y_train)
```

```python
y_pred = xgbc.predict(x_test)
```

```python
accuracy_score(y_pred,y_test)
```

# Regression

```python
from sklearn.datasets import make_regression
x,y = make_regression(n_samples = 1000,n_features =4,n_informative = 2,random_state = 42,shuffle = False)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.33,random_state =42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.ensemble import AdaBoostRegressor
reg = AdaBoostRegressor()
```

```python
reg.fit(x_train,y_train)
```

```python
y_pred = reg.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_pred,y_test)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
reg = GradientBoostingRegressor()
```

```python
reg.fit(x_train,y_train)
```

```python
y_pred = reg.predict(x_test)
```

```python
r2_score(y_pred,y_test)
```

```python
import xgboost as xb
```

```python
xgbr = xb.XGBRegressor()
```

```python
xgbr.fit(x_train,y_train)
```

```python
y_pred = xgbr.predict(x_test)
```

```python
r2_score(y_pred,y_test)
```
